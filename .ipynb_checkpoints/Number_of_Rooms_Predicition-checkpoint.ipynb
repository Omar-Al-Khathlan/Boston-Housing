# Data Partitioning

Answer:-
- I used the method 'train_test_split' as it is an included split funtion that it comes with sklearn. It takes as parameters a dataframe with the features and a dataframe for the target, then  takes a percentage input that represents the amount of allocation to the testing set. Lastly, it takes a random state input, which is used for initializing the internal random number generator, which will decide the splitting of data into train and test indices.

- The pros and cons:-
    - Pros
        - It's the standard data split method used in many Machine Learning reseach.
        - It's supported by the sklearn library.
    - Cons
        - It's a standard data split library method, which might change in future altering the results.
        - It doesn't provide everything that one might need in terms of functionality.|

```python
############################## Your code ##################################

target = data[target_name]
features = data.drop(target_name, axis=1)

x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2, random_state = 10)

# the results of the split
print("Training set has {}, ({}, 1) samples.".format(x_train.shape, y_train.shape[0]))
print("Testing  set has {}, ({}, 1) samples.".format(x_test.shape, y_test.shape[0]))
```

# Training Process

Answers:-
- An SVMs are: are supervised learning models with associated learning algorithms that analyze data used for classification and regression analysis.
- A Support Vector Regression (SVR) model will provide a better fit to this dataset compared to regular Linear Regression. Because SVRs are effective in high dimensional spaces, while Linear Regression struggles.
- The pros and cons:-
    - Pros
        - It is effective in high dimensional spaces.
        - It works really well with a clear margin of separation.
    - Cons
        - It doesnt perform well when we have a large data set. Because the required training time is higher.
        - It also doesnt perform very well, when the data set has more noise.

```python
############################## Your code ##################################

model = svm.SVR(degree=3, max_iter=-1, epsilon=0.001, C=8)
model.fit(x_train, y_train)
```

# Evaluation Process

Answers:-
- I used Cross-Validation to ensure that an independent set is used to evaluate and analyze the trained model.
- Metrics used:-
    - R Score: It compares our current model with a constant baseline and tells us how much our model is better. The constant baseline is chosen by taking the mean of the data and drawing a line at the mean.
    - Root Mean Squared Error (RMSE): It is the square root of the averaged squared difference between the target value and the value predicted by the model

```python

def evaluate_model(model_to_ev):
    
    # Training evaluation.
    y_train_predict = model_to_ev.predict(x_train)
    
    r2 = r2_score(y_train, y_train_predict)
    mse = mean_squared_error(y_train, y_train_predict)
    rmse = (np.sqrt(mse))
    
    print("\n")
    print("The model performance on training set")
    print("--------------------------------------")
    print('R2 score is {:.4f}'.format(r2))
    print('MSE is {:.4f}'.format(mse))
    print('RMSE is {:.4f}'.format(rmse))
    print("\n")
    
    # Testing evaluation.
    y_test_predict = model_to_ev.predict(x_test)
    
    r2 = r2_score(y_test, y_test_predict)
    mse = mean_squared_error(y_test, y_test_predict)
    rmse = (np.sqrt(mse))

    print("The model performance on testing set")
    print("--------------------------------------")
    print('R2 score is {:.4f}'.format(r2))
    print('MSE is {:.4f}'.format(mse))
    print('RMSE is {:.4f}'.format(rmse))
    print("\n")
```

```python
############################## Your code ##################################

evaluate_model(model)
```

# Results and Conclusions

As we can see the Support Vector Regression (SVR) model has achived an R Score of 0.7613, and a Root Mean Squared Error (RMSE) of 0.3844.

The trained model doesn't show any signs of overfitting.

```python

def graph_evaluation(model_to_ev):
    
    # Training evaluation.
    y_train_predict = model_to_ev.predict(x_train)
    
    # Visualizing the differences between actual prices and predicted values
    plt.figure(figsize=figure_size)
    plt.scatter(y_train, y_train_predict)
    plt.xlabel("Mean number of rooms")
    plt.ylabel("Predicted Mean number of rooms")
    plt.title("Mean number of rooms vs Predicted Mean number of rooms")
    plt.xlim(0,10)
    plt.ylim(0,10)
    plt.show()
    
    # Checking residuals
    plt.figure(figsize=figure_size)
    plt.scatter(y_train_predict, y_train-y_train_predict)
    plt.title("Predicted vs residuals")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.xlim(0,10)
    plt.show()

    # Checking Normality of errors
    plt.figure(figsize=figure_size)
    sns.distplot(y_train-y_train_predict)
    plt.title("Histogram of Residuals")
    plt.xlabel("Residuals")
    plt.ylabel("Frequency")
    plt.ylim(0,10)
    plt.show()

    # Testing evaluation.
    y_test_predict = model_to_ev.predict(x_test)
    
    # Visualizing the differences between actual prices and predicted values
    plt.figure(figsize=figure_size)
    plt.scatter(y_test, y_test_predict)
    plt.xlabel("Mean number of rooms")
    plt.ylabel("Predicted Mean number of rooms")
    plt.title("Mean number of rooms vs Predicted Mean number of rooms")
    plt.xlim(0,10)
    plt.ylim(0,10)
    plt.show()
    
    # Checking residuals
    plt.figure(figsize=figure_size)
    plt.scatter(y_test_predict, y_test-y_test_predict)
    plt.title("Predicted vs residuals")
    plt.xlabel("Predicted")
    plt.ylabel("Residuals")
    plt.xlim(0,10)
    plt.show()

    # Checking Normality of errors
    plt.figure(figsize=figure_size)
    sns.distplot(y_test-y_test_predict)
    plt.title("Histogram of Residuals")
    plt.xlabel("Residuals")
    plt.ylabel("Frequency")
    plt.ylim(0,10)
    plt.show()
```

```python

graph_evaluation(model)
```
